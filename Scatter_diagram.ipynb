```python
import numpy as np
import matplotlib.pyplot as plt
import math as m

x=np.array([1,2,3,4,5,6])
y=np.array([6,4,3,5,4,2])
n = len(x)

sum_xi = sum_arr(x)
sum_yi = sum_arr(y)
sum_xi_2 = sum_arr_squared(x)
sum_yi_2 = sum_arr_squared(y)
sum_xi_yi = sum_arr_arr(x,y)

r = ((n*sum_xi_yi) - (sum_xi*sum_yi))/((m.sqrt((n*sum_xi_2)-(sum_xi**2)))*(m.sqrt((n*sum_yi_2)-(sum_yi**2))))
B = (n*sum_xi_yi-sum_xi*sum_yi)/(n*sum_xi_2-sum_xi**2)
A = (sum_yi/n)-(B*(sum_xi/n))

X = np.arange(0,10,.1)
Y = A + B * X
print('A =',A)
print('B =',B)
f, ax = plt.subplots(1, 1, figsize = (18, 5))
plt.scatter(x,y, c='blue', marker='x', s=100)
for i, j in zip(x, y):
   plt.text(i, j+0.5, '({}, {})'.format(i, j),c='blue')
plt.title("Correlation coefficient (r) = %1.4f" %r )
plt.plot(X,Y,c='red')
plt.grid()
plt.show()
```

```python
def sum_arr(arr):
    sum1 =0
    for i in arr:
        sum1 = sum1 + i
    return sum1

def sum_arr_squared(arr):
    sum1 = 0
    for i in arr:
        sum1 = sum1 + (i**2)
    return sum1
def sum_arr_arr(arr1,arr2):
    sum1 = 0
    for i,j in zip(arr1,arr2):
        sum1 = sum1 + (i*j)
    return sum1
```
