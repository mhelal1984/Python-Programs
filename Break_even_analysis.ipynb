```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Q = np.linspace(0,160001,1600001)
print("Q = ",Q)
v = 20.0
p = 85.56
cf = np.full((len(Q)),902400)

a = 85.56
b = 0.000534
tr = a*Q - b*np.square(Q)

cv = Q*v
ct = []

qbe_float = cf[1]/(p-v)
qbe_int = int(qbe_float)
a = tr[qbe_int]

print(a)

for i,j in zip(cf,cv):
    z=0
    z=i+j
    ct.append(z)
```

```python
#intiate plot with size 15:10 inch
f, ax = plt.subplots(1, 1, figsize = (15, 10))
ax.set_title('Breakeven Analysis',size =25 )

plt.plot(Q,cf, linestyle='dashed',color = 'black',linewidth=2,label='Fixed Cost')
plt.plot(Q,cv, linestyle='dotted',color = 'black',linewidth=2,label='Variable Cost')
plt.plot(Q,ct,linewidth=3,color = 'red',label='Total Cost')
plt.plot(Q,tr,linewidth=3,color = 'blue',label='Total revenue')

plt.fill_between(Q, tr, ct, color='lightgrey',label='loss - profit')


intersection = np.argwhere(np.diff(np.sign(ct - tr))).flatten()
plt.plot(Q[intersection],tr[intersection],'ko',markersize=10)
print(intersection)

ax.legend(fancybox=True, shadow=True)

plt.text(Q[-1],tr[-1],"Total Revenue",size = 15)
plt.text(Q[-1],cf[-1],"Fixed cost",size = 15)
plt.text(Q[-1],ct[-1],"Total Cost",size = 15)
plt.text(Q[-1],cv[-1],"Variable Cost",size = 15)
plt.text(Q[intersection[0]+5000],tr[intersection[0]-5000],"Breakeven Point 1",size = 10,c='red')
plt.text(Q[intersection[1]+5000],tr[intersection[1]-5000],"Breakeven Point 2",size = 10,c='red')

plt.xlabel('Unit produce (Q)',size=20)
plt.ylabel('Cost & revenue',size=20)

plt.show()
```
