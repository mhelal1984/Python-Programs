First We import Libraries needed

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from sympy import *
import os
from numpy import pi as pi
```

Now definig the functions

```python
def b(v):
  """
    Calculating b(v) by running each instance of v(0:3.0 with step 0.01)

  """
  b=[]
  for i in v:
        ans = 1-(((1+math.sqrt(2))**2)/(1+(4+i**4)**0.25)**2)
        b.append(ans)
  i+=1
  return b

def dvb_dv(v):
  """
    Calculating dvb(v)/dv by running each instance of v(0:3.0 with step 0.01)

  """
  b=[]
  x = symbols('x')
  dbv_dv = diff(x*(1-(((1+math.sqrt(2))**2)/(1+(4+x**4)**0.25)**2)),x)
  for i in v:

         dvi_dv = dbv_dv.subs(x,i)
         b.append(dvi_dv)
        
  i+=1
  return b

def dvb2_dv(v):
  """
    Calculating d2vb(v)/dv2 by running each instance of v(0:3.0 with step 0.01)

  """
  b=[]
  x = symbols('x')
  dbv_dv = diff(x*(1-(((1+math.sqrt(2))**2)/(1+(4+x**4)**0.25)**2)),x,x)
  for i in v:
         dvi_dv = dbv_dv.subs(x,i)
         b.append(i*dvi_dv)
        
  i+=1
  return b
```

Now substetuting in the functions :

```python
v = np.arange(0.0,3.0,0.01)     # V-parameter values from 0 to 3 nm
b_v = []
b_v = b(v)
bdv_v = dvb_dv(v)
bdv2_v = dvb2_dv(v)
```

Plotting the functions :

```python
fig, ax = plt.subplots(figsize = (15,8 ))
ax.set_title('Mohamed Helal Sec#4',size =25 )

plt.plot(v,b_v, linestyle='dashed',linewidth=2,label='b(v)')
plt.plot(v,bdv_v,linewidth=2,label='d(vb)/dv')
plt.plot(v,bdv2_v,linewidth=2,label='vd2(vb)/dv2')

ax.legend(fancybox=True, shadow=True)
plt.xlabel('V')
plt.ylabel('b(V)')
plt.xlim([0, 2.99])
plt.show()
```
