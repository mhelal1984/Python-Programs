```python
import matplotlib.pyplot as plt
import numpy as np
import math
import os
```

```python
n1 = 1.5
n2 = n1 * 0.99
delta = (n1-n2)/n1
a = 4
g=[1,2,5,np.inf]
r = np.arange(0,6.0,.01)
n_r= []
y = []
```

```python
def n(r) :                                                             
    y = []
    for j in range(0,len(g)):
        g_element = g[j]
        n_r = []
        for i in range(0,len(r)):                                           
            
            if 4>r[i]>0:                                                    
                x = (math.sqrt((n1**2)-(n2**2))*(math.sqrt(1-((r[i])/a)**g_element)))       
                n_r.append(x)                                                                                             
            else:                                                           
                x = 0                                                      
                n_r.append(x)                                                
        i+=1                                                               
        y.append(n_r)
    j+=1
    return y
N_R = n(r)
```

```python
fig, ax = plt.subplots()
ax.set_title('Refractive Index Profile')
line1, = ax.plot(r, N_R[0], lw=2,color='black',label = 'g = 1')
line2, = ax.plot(r, N_R[1], lw=2,color='green',label = 'g = 2')
line3, = ax.plot(r, N_R[2], lw=2,color='red',label = 'g = 5')
line4, = ax.plot(r, N_R[3], lw=1 ,color='orange',label = 'g = >>')
leg = ax.legend(fancybox=True, shadow=False)

lines = [line1, line2, line3, line4]
plt.xlabel('Radius')
plt.ylabel('NA(r)')
plt.xlim([0, 4.2])
lined = {}  # Will map legend lines to original lines.
for legline, origline in zip(leg.get_lines(), lines):
    legline.set_picker(True)  # Enable picking on the legend line.
    lined[legline] = origline
plt.show()
```
